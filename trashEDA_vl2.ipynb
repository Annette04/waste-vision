## Анализ размера изображений

```python
from PIL import Image
from io import BytesIO
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt

# Функция для получения размеров изображений
def get_image_size(img_dict):
    img = Image.open(BytesIO(img_dict["bytes"]))
    return img.size

# Получаем размеры изображений
df["width"], df["height"] = zip(*df["image"].apply(get_image_size))

# Подсчитаем частоту уникальных размеров
size_counter = Counter(zip(df["width"], df["height"]))
most_common_sizes = size_counter.most_common(5)

# Выводим топ-5 самых частых размеров
print("Топ 5 самых частых размеров изображений:")
for size, count in most_common_sizes:
    print(f"Размер: {size}, Количество: {count}")

# Визуализация распределения ширины и высоты
plt.figure(figsize=(10, 5))
sns.histplot(df["width"], bins=20, color="blue", label="Ширина")
sns.histplot(df["height"], bins=20, color="red", label="Высота")
plt.legend()
plt.title("Распределение размеров изображений")
plt.show()
```

### Соотношение сторон в основном сосредоточенно в значениях 416 на 416. Есть выбросы

- Размер: (416, 416), Количество: 17365
- Размер: (4032, 3024), Количество: 2163
- Размер: (3264, 2448), Количество: 364

исходя из результатов, попробуем привести все изображения к кваратному виду 416 на 416, путем их растягивания



```python
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# Функция для изменения размера и сохранения исходного размера
def resize_and_track_original(img_dict, target_size=(416, 416)):
    img = Image.open(BytesIO(img_dict["bytes"]))
    original_size = img.size
    img_resized = img.resize(target_size)

    buffer = BytesIO()
    img_resized.save(buffer, format=img.format)
    img_dict["bytes"] = buffer.getvalue()
    img_dict["original_size"] = original_size

    return img_dict

# Применяем Resize ко всем изображениям
df["image"] = df["image"].apply(resize_and_track_original)

# Фильтруем изображения, которые изначально были другого размера
different_size_images = df["image"].apply(lambda x: x["original_size"] != (416, 416))

# изображения для демонстрации изменений
sample_images = df[different_size_images].head(6)["image"]

# Визуализация
plt.figure(figsize=(15, 10))
for i, img_dict in enumerate(sample_images):

    img_original = Image.open(BytesIO(img_dict["bytes"]))
    img_original = img_original.resize(img_dict["original_size"])  # Восстанавливаем исходный размер

    img_resized = Image.open(BytesIO(img_dict["bytes"]))

    plt.subplot(2, 6, i+1)
    plt.imshow(img_original)
    plt.axis('off')
    plt.title(f"Исходный\n{img_dict['original_size']}")

    plt.subplot(2, 6, i+7)
    plt.imshow(img_resized)
    plt.axis('off')
    plt.title("После Resize (416x416)")

plt.suptitle("Сравнение изображений до и после Resize")
plt.show()
```

Изменяем размер изображений путем их деформации до размера 416 на 416. Выше приведены примеры измененных изображений
